# 变量可视化展示

```python
class df_visualization():
    def __init__(self, df):
        try:
            if not isinstance(df,pd.DataFrame):
                raise TypeError('TypeError')
            else:
                self.df=df
        except TypeError:
            print("数据类型错误！")

    def plot_categorical_variable(self, var, color):

        
        order=self.df[var].value_counts().index
        cnt=len(order)

        # 图片设置
        plt.figure(figsize=(16, 8))
        plt.suptitle(var+' Distribution', fontweight='heavy', 
                        fontsize='16', fontfamily='sans-serif', color=black_grad[0])

        # 饼图
        plt.subplot(1, 2, 1)
        plt.title('Pie Chart', fontweight='bold', fontsize=14,
                fontfamily='sans-serif', color=black_grad[0])
        plt.pie(self.df[var].value_counts(), labels=order, colors=color, pctdistance=0.7,
                autopct='%.2f%%', wedgeprops=dict(alpha=0.8, edgecolor=black_grad[1]),
                textprops={'fontsize':12})
        centre=plt.Circle((0, 0), 0.45, fc='white', edgecolor=black_grad[1])
        plt.gcf().gca().add_artist(centre)

        # 直方图
        countplt = plt.subplot(1, 2, 2)
        plt.title('Histogram', fontweight='bold', fontsize=14, 
                fontfamily='sans-serif', color=black_grad[0])
        ax = sns.countplot(x=var, data=self.df, palette=color, order=order,
                        edgecolor=black_grad[2], alpha=0.85)
        for rect in ax.patches:
                ax.text (rect.get_x()+rect.get_width()/2, 
                        rect.get_height()+4,rect.get_height(), 
                        horizontalalignment='center', fontsize=10, 
                        bbox=dict(facecolor='none', edgecolor=black_grad[0], 
                                linewidth=0.25, boxstyle='round'))

        plt.xlabel(var, fontweight='bold', fontsize=11, fontfamily='sans-serif', 
                color=black_grad[1])
        plt.ylabel('Total', fontweight='bold', fontsize=11, fontfamily='sans-serif', 
                color=black_grad[1])
        plt.xticks(range(cnt), order)
        plt.grid(axis='y', alpha=0.4)
        countplt

    def plot_numerical_variable(self, var, color):
        
        fig=plt.figure(figsize=(12, 12))
        fig.suptitle(var+' Column Distribution', fontweight='bold', fontsize=16, 
                        fontfamily='sans-serif', color=black_grad[0])
        fig.subplots_adjust(top=0.9)

        # 直方图
        ax_1=fig.add_subplot(2, 2, 2)
        plt.title('Histogram Plot', fontweight='bold', fontsize=14, 
                fontfamily='sans-serif', color=black_grad[1])
        sns.histplot(data=self.df, x=var, kde=True, color=color)
        plt.xlabel('Total', fontweight='regular', fontsize=11, 
                fontfamily='sans-serif', color=black_grad[1])
        plt.ylabel(var, fontweight='regular', fontsize=11, fontfamily='sans-serif', 
                color=black_grad[1])

        # 正态Q-Q图
        ax_2=fig.add_subplot(2, 2, 4)
        plt.title('Q-Q Plot', fontweight='bold', fontsize=14, 
                fontfamily='sans-serif', color=black_grad[1])
        qqplot(self.df[var], fit=True, line='45', ax=ax_2, markerfacecolor=color, 
                markeredgecolor=color, alpha=0.6)
        plt.xlabel('Theoritical Quantiles', fontweight='regular', fontsize=11, 
                fontfamily='sans-serif', color=black_grad[1])
        plt.ylabel('Sample Quantiles', fontweight='regular', fontsize=11, 
                fontfamily='sans-serif', color=black_grad[1])

        # 箱型图
        ax_3=fig.add_subplot(1, 2, 1)
        plt.title('Box Plot', fontweight='bold', fontsize=14, fontfamily='sans-serif', 
                color=black_grad[1])
        sns.boxplot(data=self.df, y=var, color=color, boxprops=dict(alpha=0.8), linewidth=1.5)
        plt.ylabel(var, fontweight='regular', fontsize=11, fontfamily='sans-serif', 
                color=black_grad[1])

        plt.show()

    def summary_numerical_variable(self,var_lst):
        Skewness_lst=[]
        Kurtosis_lst=[]
        for var in var_lst:
            Skewness_lst.append(self.df[var].skew(axis = 0, skipna = True))
            Kurtosis_lst.append(self.df[var].kurt(axis = 0, skipna = True))
        SK_table=pd.DataFrame({'Skewness':Skewness_lst, 'Kurtosis':Kurtosis_lst})
        SK_table.index=var_lst
        return SK_table


df_visual=df_visualization(df)
```

# 预处理

```python
# 建立哑变量
df = pd.get_dummies(df, columns = lst)

df.head()
```

```python
# 分离目标变量
x = df.drop(['target'], axis=1)
y = df['target']
```

```python
# 分离训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

# 模型训练

```python
class classifier():
    def __init__(self,x_train, x_test, y_train, y_test,name,clf):
        try:
            self.x_train=x_train
            self.x_test=x_test
            self.y_train=y_train
            self.y_test=y_test
            self.name=name
            self.clf=clf
            clf.fit(x_train, y_train)
            self.y_pred = clf.predict(x_test)
            self.Acc = accuracy_score(self.y_test, self.y_pred)
            self.Pre = precision_score(self.y_test, self.y_pred)
            self.Recall = recall_score(self.y_test, self.y_pred)
            self.F1 = f1_score(self.y_test, self.y_pred)
            self.auc = roc_auc_score(self.y_test, self.y_pred)
        except ValueError:
            print("传入参数维度可能不匹配！")
    # 准确度
    def print_Acc(self):
        print('.:. Logistic Regression Accuracy:'+'\033[1m {:.2f}%'.format(self.Acc*100)+' .:.')

    # 分类效果
    def clf_report(self):
        print('\n\033[1m'+'.: Classification Report'+'\033[0m')
        print('*' * 25)
        print(classification_report(self.y_test, self.y_pred))

    # 可视化评价
    def visual_evaluation(self):
        print('\n\033[1m'+'.: Performance Evaluation'+'\033[0m')
        print('*' * 26)
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))

        # 混淆矩阵
        matrix = ConfusionMatrix(self.clf, ax=ax1, cmap='PuRd',
                                    title=self.name+' Confusion Matrix')
        matrix.fit(self.x_train, self.y_train)
        matrix.score(self.x_test, self.y_test)
        matrix.finalize()

        # -Roc曲线
        rocauc = ROCAUC(self.clf, classes=['False', 'True'], ax=ax2,
                        title=self.name+' ROC AUC Plot')
        rocauc.fit(self.x_train, self.y_train)
        rocauc.score(self.x_test, self.y_test)
        rocauc.finalize()

        # 学习曲线
        lc = LearningCurve(self.clf, ax=ax3, title=self.name+' Learning Curve')
        lc.fit(self.x_train, self.y_train)
        lc.finalize()

        # Precision-Recall曲线
        prcurve = PRCurve(self.clf, ax=ax4, ap_score=True, iso_f1_curves=True,
                                        title=self.name+' Precision-Recall Curve')
        prcurve.fit(self.x_train, self.y_train)
        prcurve.score(self.x_test, self.y_test)
        prcurve.finalize()

        plt.tight_layout()
    
    def model_performance(self):
        self.print_Acc()
        self.clf_report()
        self.visual_evaluation()
```

## Logistic Regression

```python
# 寻找最优参数
param = {'penalty':['l1','l2'],'solver':['liblinear','lbfgs','sag','newton-cg']}
grid = GridSearchCV(LogisticRegression(),param_grid=param,cv=10)
grid.fit(x_train,y_train)
print(grid.best_params_) 
```

```python
lr=LogisticRegression(random_state=1, solver='liblinear', penalty='l2')
LRClassifier=classifier(x_train, x_test, y_train, y_test, 'Logistic Regression',lr)
LRClassifier.model_performance()
```

## KNN

```python
#寻找最优邻居个数
scorel=[]
for k in range(1, 10):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    score = knn.score(x_test,y_test)
    scorel.append(score) 
plt.figure(figsize=[8,5]) 
plt.plot(range(1,10),scorel) 
plt.xlabel('Number of K')
plt.ylabel('Score')
plt.title('Find the best K for KNN', fontweight='bold', fontsize=14,
                fontfamily='sans-serif', color=black_grad[0])
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
KNNClassifier=classifier(x_train, x_test, y_train, y_test, 'KNN', knn)
KNNClassifier.model_performance()
```

## 决策树

```python
# 寻找最优参数，这里主要调整数的深度
param = {'max_depth':[2,3,4,5],'min_samples_leaf':[3,5,7,10]}
grid = GridSearchCV(DecisionTreeClassifier(),param_grid=param,cv=5)
grid.fit(x_train,y_train)
print(grid.best_params_) 
```

```python
dtc = DecisionTreeClassifier(max_depth=3, min_samples_leaf=5, criterion='entropy', min_samples_split=5,
                                       splitter='random', random_state=1)
DTCClassifier=classifier(x_train, x_test, y_train, y_test, 'Decision Tree', dtc)
DTCClassifier.model_performance()
```

## 随机森林

```python
# 寻找最优的n_estimators
scorel = [] 
for i in range(0,500,10):    
	rfc = RandomForestClassifier(n_estimators=i, n_jobs=-1, random_state=1)    
	score = cross_val_score(rfc,x_train,y_train,cv=10).mean()    		
	scorel.append(score) 
plt.figure(figsize=[8,5]) 
plt.plot(range(1,501,10),scorel) 
plt.xlabel('Number of n_estimators')
plt.ylabel('Score')
plt.title('Find the best n_estimators for Random Forest', fontweight='bold', fontsize=14,
                fontfamily='sans-serif', color=black_grad[0])
plt.show()
```

```python
# 在120附近继续搜索
scorel = [] 
for i in range(110,130):    
	rfc = RandomForestClassifier(n_estimators=i, n_jobs=-1, random_state=1)    
	score = cross_val_score(rfc,x_train,y_train,cv=10).mean()    		
	scorel.append(score) 
plt.figure(figsize=[8,5]) 
plt.plot(range(110,130),scorel) 
plt.xlabel('Number of n_estimators')
plt.ylabel('Score')
plt.title('Find the best n_estimators for Random Forest', fontweight='bold', fontsize=14,
                fontfamily='sans-serif', color=black_grad[0])
plt.show()
```

```python
rf = RandomForestClassifier(n_estimators=120, random_state=1, max_leaf_nodes=20, min_samples_split=15)
RFClassifier=classifier(x_train, x_test, y_train, y_test, 'Random Forest', rf)
RFClassifier.model_performance()
```
